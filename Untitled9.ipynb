```python
import requests
from bs4 import BeautifulSoup

# URL of the website you want to scrape
url = 'https://www.scrapethissite.com/pages/simple/'

# Send a GET request to the website
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # Example: Extracting all instances of a specific tag
    for element in soup.find_all('h1'):
        print(element.text)
else:
    print(f"Failed to retrieve the webpage: Status code {response.status_code}")
```

```python
import requests

url = 'https://www.scrapethissite.com/pages/simple/'
response = requests.get(url)
html_content = response.text
```

```python
print(response.text)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html_content, 'html.parser')
```

```python
# Load the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Find all country divs
countries_divs = soup.find_all('div', class_='country')

# Iterate through each country div and extract information
for country_div in countries_divs:
    country_name = country_div.find('h3', class_='country-name').text.strip()
    capital = country_div.find('span', class_='country-capital').text.strip()
    population = country_div.find('span', class_='country-population').text.strip()
    area = country_div.find('span', class_='country-area').text.strip()

    # You can then print out or store this information
    print(f"Country: {country_name}, Capital: {capital}, Population: {population}, Area: {area} km²")
```

```python
table = soup.find('table', class_='your-table-class')  # Replace 'your-table-class' with the actual class name of the table.
rows = table.find_all('tr')

data = []
for row in rows:
    cols = row.find_all(['td', 'th'])
    cols = [ele.text.strip() for ele in cols]
    data.append(cols)
```
