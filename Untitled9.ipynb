```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
```

```python
cars = pd.read_csv('/content/cars_ds_final.csv')
```

```python
cars.describe()
```

```python
null_values_sum = cars.isnull().sum()
print(null_values_sum)
```

```python
cars.dtypes
```

```python
cars.columns
```

```python
cars.rename(columns = str.lower,inplace=True)
cars.columns
```

```python
cars.make.isna().sum()
```

```python
cars['make'].value_counts()
```

```python
cars['make'] = cars['make'].replace('Land Rover Rover','Land Rover')
```

```python
cars['ex-showroom_price'] = cars['ex-showroom_price'].str.replace('Rs. ', '').str.replace(',', '').astype(float)
```

```python
import seaborn as sns
```

```python
plt.hist(cars['ex-showroom_price'], bins=20, edgecolor='k')
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(data=cars, x='ex-showroom_price', kde=True)
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
manufacturer_counts = cars['make'].value_counts()

plt.figure(figsize=(10, 8))
manufacturer_counts.plot(kind='bar')
plt.title('Count of Cars by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Count')
plt.xticks(rotation=70)
```

```python
bins = [0, 500000, 1000000, 1500000, 2000000, 5000000, 10000000 ,20000000,30000000]

# Labels for the bins
labels = ['0-5 Lakh', '5-10 Lakh', '10-15 Lakh', '20-50 Lakh', '50 Lakh+','1 crore','2 crore','3 crore']

# Segmenting 'Ex-Showroom_Price' into ranges
cars['Price_Range'] = pd.cut(cars['ex-showroom_price'], bins=bins, labels=labels, include_lowest=True)

# Check the distribution across the new price ranges
print(cars['Price_Range'].value_counts())
```

```python
sns.boxplot(x='make', y='Price_Range', data=cars)
plt.xticks(rotation=80)

plt.title('Numeric Variable Distribution by Categorical Variable')
plt.show()
```

```python
cars['make'].value_counts()
```

```python
N = 10
top_N_makes = cars['make'].value_counts().nlargest(N).index

# Create a new column where only top N categories are retained and others are labeled as 'Other'
cars['Top_Make'] = cars['make'].apply(lambda x: x if x in top_N_makes else 'Other')

# Verify changes
print(cars['Top_Make'].value_counts())
```

```python
Top_make = cars['Top_Make'].value_counts()

# Plot a pie chart
plt.figure(figsize=(10, 8))
Top_make.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Contribution of Top 10 Makes Including \'Other\'')
plt.ylabel('')  # Hide the y-label as it's not needed for pie charts
plt.show()
```

```python
cars['fuel_type'].value_counts()
```

```python
bins = [0, 500000, 1000000, 1500000, 2000000, 5000000, 10000000 ,20000000,30000000]

# Labels for the bins
labels = ['0-5 Lakh', '5-10 Lakh', '10-15 Lakh', '20-50 Lakh', '50 Lakh+','1 crore','2 crore','3 crore']

# Segmenting 'Ex-Showroom_Price' into ranges
cars['Price_Range'] = pd.cut(cars['ex-showroom_price'], bins=bins, labels=labels, include_lowest=True)

# Check the distribution across the new price ranges
print(cars['Price_Range'].value_counts())
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='fuel_type', y='Price_Range', data=cars)
plt.xticks(rotation=45)  # Rotate the x-axis labels to avoid overlap
plt.title('Price Distribution by Fuel Type')
plt.show()
```

```python
corr_matrix = cars.select_dtypes(include=['float64', 'int64']).corr()

# Visualizing the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numeric Variables')
plt.show()
```

```python
numeric_columns = ['ex-showroom_price', 'displacement', 'cylinders', 'city_mileage']  # Example columns
cars_numeric_subset = cars[numeric_columns]

# Creating the pair plot
sns.pairplot(cars_numeric_subset)
plt.show()
```
