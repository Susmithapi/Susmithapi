```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
matches_per_year = matches['season'].value_counts().sort_index()

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=matches_per_year.index, y=matches_per_year.values, palette="Blues_d")
plt.title('Number of Matches Played Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.show()
```

```python
runs_by_batsman = deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 6))
runs_by_batsman.plot(kind='bar')
plt.title('Top 10 Run Scorers')
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.xticks(rotation=45)
plt.show()
```
